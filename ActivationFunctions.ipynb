## Logistic Sigmoid　(シグモイド関数)

**標準シグモイド関数**<br>
入力された実数値を0から1の間の確率として表現するS字型の非線形関数。特にロジスティック回帰の出力や、かつてのニューラルネットワークの活性化関数として使われる。
$$
f(x) = \frac{1}{1 + e^{-x}}
$$

* この関数は $(-\infty, \infty) \rightarrow(0, 1)$ の短調増加連続関数である。$y=0, y=1$ を漸近線として持っていて，確率値のような値域なのでMLP(Multi-Layer Perceptron: 多層パーセプトロン)では確率を出したい場合に使われることもある。
* 出力が0〜1に収まるため確率解釈に適しているが、入力が極端な値だと勾配が非常に小さくなる勾配消失問題があり、現在の深層学習ではReLUなどが主流となっている。

```python
# sigmoid関数
sigmoid = lambda x: 1/(1+np.exp(-x))
sigmoid_prime = lambda x: sigmoid(x)*(1-sigmoid(x))
```

```python
# 可視化
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)

x = np.linspace(-10,10,200)

plt.title("sigmoid function")
plt.xlabel("input $x$")
plt.ylabel("output")
plt.plot(x,sigmoid(x),label="forward")
plt.legend()

plt.subplot(1,2,2)
plt.title("sigmoid function")
plt.xlabel("input $x$")
plt.ylabel("output")
plt.plot(x,sigmoid(x), label="forward")
plt.plot(x,sigmoid_prime(x),label="backward")
plt.legend()
plt.show()
```
